```python
# importing full file
import pandas as pd

file_path = r"C:\Users\sneha\3D Objects\Desktop\career\Airline_performance_analysis\Dataset\flights_clean.csv"

chunk_size = 50000

for i, chunk in enumerate(pd.read_csv(file_path, chunksize=chunk_size)):
    chunk.to_sql(
        name="flights",
        con=engine,
        if_exists="append",
        index=False,
        method="multi"
    )
    
    print(f"Chunk {i+1} loaded")
    
print("Full dataset imported successfully!")
```
